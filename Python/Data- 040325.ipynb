## Normality Test

```python
from scipy.stats import shapiro
```

```python
numeric_columns = ['Interpersonal Skills Rating', 'Decision-Making Skills Rating', 
                   'Task Completion Rates', 'Sales Revenue Generated']
shapiro_results = {}

for column in numeric_columns:
    stat, p_value = shapiro(df[column])
    shapiro_results[column] = round(p_value, 3)
    
shapiro_results
```

## Square Root Transformation

```python
def sqrt_transformation(data, column_name):
    
    data[f'{column_name}_sqrt'] = np.sqrt(data[column_name])
    stat, p_value = shapiro(data[f'{column_name}_sqrt'])
    distribution = sns.kdeplot(data[f'{column_name}_sqrt'], fill=True)
    
    print(distribution)
    print('p-value: ', p_value)
    
sqrt_transformation(df, 'Sales Revenue Generated')
```

## Boxcox Transformation

```python
from scipy.stats import boxcox
```

```python
def boxcox_transformation(data, column_name):
    transformed_data, _ = boxcox(data[column_name])   
    data[f'{column_name}_boxcox'] = transformed_data
    stat, p_value = shapiro(data[f'{column_name}_boxcox'])
    kdeplot = sns.kdeplot(data[f'{column_name}_boxcox'], fill=True)

    print(kdeplot)
    print('P value: ',p_value)    
```

```python
boxcox_transformation(df, 'Sales Revenue Generated')
```

## Yeojohnson Transformation

```python
from scipy.stats import yeojohnson
```

```python
def yeojohnson_transformation(data, column_name):
    transformed_data, _ = yeojohnson(data[column_name])
    data[f'{column_name}_yeojohnson'] = transformed_data
    stat, p_value = shapiro(data[f'{column_name}_yeojohnson'])
    kdeplot = sns.kdeplot(data[f'{column_name}_yeojohnson'], fill=True)
    print(kdeplot)
    print('P value: ', p_value)
```

```python
yeojohnson_transformation(df, 'Sales Revenue Generated')
```

## One Sample T-Test

```python
from scipy import stats
```

```python
hypothesis_mean = 68
```

```python
t_statistic, p_value = stats.ttest_1samp(df['Task Completion Rates'], hypothesis_mean)

print("P-value:", p_value)
```

```python
print("t_statistic:", t_statistic)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the average purchase amount and 68.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the average purchase amount and 68.")
```

## Independent sample t-test

```python
churned = df.query('churned== "Yes"')['Task Completion Rates']
existing = df.query('churned == "No"')['Task Completion Rates']
t_statistic, p_value = stats.ttest_ind(churned, existing)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in average purchase amount between churned andexisting customers.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in average purchase amount between churnedand existing customers.")
```

## One way ANOVA

```python
df1 = pd.read_excel("additional_data.xlsx")
```

```python
df1.head()
```

```python
Chicago = df1.query('City == "Chicago"')['Frequency_of_Purchases']
Houston = df1.query('City == "Houston"')['Frequency_of_Purchases']
Los_angeles = df1.query('City == "Los Angeles"')['Frequency_of_Purchases']
t_statistic, p_value = stats.f_oneway(Chicago, Houston, Los_angeles )
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print('Reject the null hypothesis. There is a significant difference in average frequency of purchases among the customersfrom different cities.')
else:
    print('Fail to reject the null hypothesis. There is no significant difference in average frequency of purchases among thecustomers from different cities.')
```
