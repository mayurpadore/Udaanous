## Problem 1: 



```python
def compound(P,r,n,t,A):
    A = P*(1+r/n)**(n*t)
    
    print("Compound Interest is :", A)

P = int(input(" Principle :"))
r = int(input(" Rate of Interest :"))
n = int(input(" number of times interest is compounded per year :"))
t = int(input(" Time in Years :"))


compound(P,r,n,t,A)
```

## Problem 2: 



```python
def installent(P,r,n,EMI):
    r = r / (12 * 100)  
    EMI = P*r*((1+r)**n)/((1+r)**n-1) 
    
    print("EMI for the Loan is ", EMI)

P = int(input(" Principle :"))
r = int(input(" Rate of Interest :"))
n = int(input(" number of times interest is compounded per year :"))

installent(P,r,n,EMI)
```

## Problem 3:



```python
def index(w,h,BMI):
    BMI = w/(h**2)
    
    if BMI < 18.5:
        print("Underweight")
    elif BMI >= 18.5 and BMI <25:
        print("Normal Weight")
    elif BMI >= 25 and BMI < 30:
        print("Overweight")
    else:
        print("Obesity")
    

w = int(input(" Weight in kilogram :"))
h = float(input(" Height in meter :"))


index(w,h,BMI)
```
